```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import glob
import random
import cv2
from sklearn.model_selection import train_test_split
import tensorflow as tf


#NN
from tensorflow.keras.preprocessing.image import img_to_array, ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Activation, Dense, Dropout, Flatten
from tensorflow.keras import backend as k
from tensorflow.keras.optimizers import Adam
```

```python
#initializing the hyperparameters:
epochs  = 100
img_dimension = (80,80,3)
batch = 64
```

```python
#initialising images and labels for training
data = [] #There are 2307 image in man and woman combined. 1173 man images,  1134 woman images
labels = []


image_files=[f for f in glob.glob(r'./gender_dataset_face' + "/**/*", recursive=True) if not os.path.isdir(f)]
random.shuffle(image_files) #this is done in order to change the order of man and woman files for learning better
```

```python
#reading the image files :
for img in image_files:
    image  = cv2.imread(img)
    image = cv2.resize(image,(img_dimension[0],img_dimension[1])) #to keep all the images of the same dimensions.
    image = img_to_array(image)
    data.append(image)
    
    #assigning the labels for the images
    label = img.split(os.path.sep)[-2] #-2 indicates the the folder name 'man' or 'woman'
    
    if label == 'man':
        label = 1
    else:
        label = 0
    labels.append([label]) #the labels stored in 2D arrays
```

```python
data = np.array(data, dtype = 'float')
data = data/255.0 #feature scaling
labels = np.array(labels)
labels = to_categorical(labels, num_classes = 2) #[[0,1],[0,1]] first position is man, hence 1 indicates that label
#hot encoded labels after to_categorical are always set in [0,1] system
```

```python
xtrain ,xtest, ytrain, ytest = train_test_split(data,labels, test_size = 0.2, random_state = 0)
```

```python
#Function for building a neural network
def network(width, height, depth, classes):
    model  = Sequential()
    input_dim = (height, width, depth)
    chan_dim = -1
    
    if k.image_data_format == "channels_first":
        chan_dim = 1
    
    #Constructing the model. notice that no of conv filters increase and pool_size decreases as we construct layers. 
    model.add(Conv2D(32,(3,3), padding = 'same', input_shape = input_dim))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis = chan_dim))
    model.add(MaxPooling2D(pool_size = (3,3)))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(64,(3,3), padding = 'same'))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis = chan_dim))
    
    model.add(Conv2D(64, (3,3), padding = 'same',))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis = chan_dim))
    model.add(MaxPooling2D(pool_size = (2,2)))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(128, (3,3), padding = 'same'))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis = chan_dim))
    
    model.add(Conv2D(128, (3,3), padding = 'same'))
    model.add(Activation("relu"))
    model.add(BatchNormalization(chan_dim))
    model.add(MaxPooling2D(pool_size = (2,2)))
    model.add(Dropout(0.25))
    
    #Fully connected Layer
    model.add(Flatten())
    model.add(Dense(1024))
    model.add(Activation("relu"))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    
    model.add(Dense(classes))
    model.add(Activation('sigmoid'))
    
    return model
```

```python
#calling for model function 'network' 
model  = network(img_dimension[0],img_dimension[1],img_dimension[2], 2)
#model.summary()
```

```python
#Image Augmentation: Produces different variations of training data for the CNN to learn better
aug = ImageDataGenerator(horizontal_flip = True, 
                         fill_mode = 'nearest', shear_range = 0.2, 
                         zoom_range = 0.2, rotation_range =25, width_shift_range = 0.1, height_shift_range = 0.1)
```

```python
optimizer = Adam(lr = 0.001)

model.compile(loss = "binary_crossentropy", optimizer = optimizer, metrics = ["accuracy"])

#training the model on the images

trained = model.fit_generator(aug.flow(xtrain,ytrain, batch_size = batch), 
                              validation_data = (xtest,ytest),steps_per_epoch = len(xtrain)//batch, 
                                       epochs = epochs,verbose = 1)
                              
                              
#model.save("model.h5")                            
```

```python
#Visualizations
plt.figure()
plt.plot(np.arange(0,epochs), trained.history["loss"], label = "train_loss")
plt.plot(np.arange(0,epochs), trained.history["val_loss"], label  = "val_loss")
plt.title("Training loss vs Validation Loss")
plt.xlabel("epochs")
plt.ylabel("Loss")
plt.legend(loc = "upper right")
plt.show()

plt.figure()
plt.plot(np.arange(0,epochs), trained.history["accuracy"], label =  "train_acc")
plt.plot(np.arange(0,epochs), trained.history["val_accuracy"], label = "validation acc")
plt.title("Training accuracy vs Validation accuracy")
plt.xlabel("epochs")
plt.ylabel("Accuracy")
plt.legend(loc = "upper right")
plt.show()
#plt.savefig('plt.png')
```
